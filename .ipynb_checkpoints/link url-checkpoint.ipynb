```python
import requests
import time
```

```python
URL ="http://127.0.0.1:8000/api/device/d000"
```

```python
while True:
    res = requests.get(URL)
    data= res.json()
    
    if data['light']==1 and data['fan']==1:
        print("light and fan on")
    elif data['light']==1 and data['fan']==0:
        print("only light on")
    elif data['light']==0 and data['fan']==1:
        print("only fan on")
    else:
        print("all off")
    time.sleep(1)
```
